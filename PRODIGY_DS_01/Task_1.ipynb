# Extraction of top 10 countries with respect to total population

```python
# filter data for total population
total_population_data=df[df['Indicator Code']=='SP.POP.TOTL']

# Sort data based on the total population for 2024
total_population_sorted = total_population_data.sort_values(by="2024",ascending=False)

# Get the top ten countries with the highest total population for 2024
total_top_ten_countries = total_population_sorted.head(10)

print("Top ten countries of total population \n")
print(total_top_ten_countries[['Country Code']])
```

# Bar Plot

## Top ten countries of total population in year 2024 and 2010

```python
# Create a bar plot 2024
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=total_top_ten_countries,palette="magma")
plt.title("Top ten countries of total population (2024)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create a bar plot for 2010
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2010",y="Country Code",data=total_top_ten_countries,palette="magma")
plt.title("Top ten countries of total population (2010)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# Extraction of bottom 10 countries with respect to total population

```python
# Sort data based on the total population for 2024
total_population_sorted = total_population_data.sort_values(by="2024",ascending=True)

# Get the top ten countries with the highest total population for 2024
total_bottom_ten_countries = total_population_sorted.head(10)

print("Bottom ten countries oftotal population \n")
print(total_bottom_ten_countries[['Country Code']])
```

## Extraction of top 10 countries with highest male population

```python
# Filter data for male population 
male_population_data = df[df['Indicator Code']=='SP.POP.TOTL.MA.IN']

# Sort data based on the male population for 2024
male_population_sorted = male_population_data.sort_values(by='2024',ascending = False)

# Get the top 10 countries with the highest male population for 2024
male_top_ten_countries = male_population_sorted.head(10)
print('Top ten countries of male population')
print(male_top_ten_countries[['Country Code']])
```

## Extraction of top 10 countries with highest female population

```python
# Filter data for female population 
female_population_data = df[df['Indicator Code']=='SP.POP.TOTL.FE.IN']

# Sort data based on the male population for 2024
female_population_sorted = female_population_data.sort_values(by='2024',ascending = False)

# Get the top 10 countries with the highest female population for 2024
female_top_ten_countries = female_population_sorted.head(10)
print('Top ten countries of female population')
print(female_top_ten_countries[['Country Code']])
```

## Top ten countries with highest male and female population in 2024

```python
# Create a bar plot for Top ten countries of male population 2024
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2024",y="Country Code",data=male_top_ten_countries,palette="viridis")
plt.title("Top ten countries of male population (2024)",fontsize=10)
plt.xlabel("Male population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create a bar plot for Top ten countries of female population 2024
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2010",y="Country Code",data=female_top_ten_countries,palette="viridis")
plt.title("Top ten countries of female population (2024)",fontsize=10)
plt.xlabel("Female population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# Stacked Bar Plot

## Top ten countries with male and female population

```python
# Merge male and female population data on 'Country Code'
merge_data = pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
# merged data
# Calculate the total population for each country (male + female)
merge_data["Total population"] = merge_data["2024_male"] + merge_data["2024_female"]
```

```python
merge_data.head()
```

```python
# sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total population",ascending=False)
```

## Select the top 10 countries with the highest total population

```python
top_10_countries = sorted_data.head(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2024_female",data=top_10_countries,color="#F5004F",label="female population")
sns.barplot(x="Country Code",y="2024_male",data=top_10_countries,bottom=top_10_countries["2024_female"],color="#000957",label="male population")
plt.xlabel("Country")
plt.ylabel("Population (In Billions)")
plt.title("Top 10 Countries' Gender Distribution - 2024",fontsize=20)
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

## Bottom 10 countries with make and female population (2024)

```python
# Select the bottom 10 countries with th highest population 
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2024_female",data=bottom_10_countries,color="#ffbe62",label="female population")
sns.barplot(x="Country Code",y="2024_male",data=bottom_10_countries,bottom=bottom_10_countries["2024_female"],color="#000000",label="male population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Bottom 10 Countries' Gender Distribution - 2024",fontsize=20)
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
